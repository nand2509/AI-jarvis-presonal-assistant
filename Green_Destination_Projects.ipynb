```python
import pandas as pd

# Assuming you have already read the CSV file into a DataFrame
df = pd.read_csv('greendestination.csv')

# Print unique values in the 'Attrition' column before mapping
print(df['Attrition'].unique())

# Convert 'Attrition' column into integers (1 and 0)
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else (0 if x == 'No' else -1))

# Print the 'Attrition' column after conversion
print(df['Attrition'])
```

```python
df.head(5)
```

```python
print(df['Attrition'])
```

```python
# prompt: graph  between age and attrition

import matplotlib.pyplot as plt

# Create a bar chart of age vs. attrition
df.groupby('Age')['Attrition'].mean().plot(kind='bar')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Age")

# Show the plot
plt.show()
```

```python
df.groupby('BusinessTravel')['Attrition'].mean().plot(kind='bar')

# Add labels and title
plt.xlabel("Business Travel")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Business Travel")

# Show the plot
plt.show()
```

```python
df.groupby('DistanceFromHome')['Attrition'].mean().plot(kind='bar')

# Add labels and title
plt.xlabel("DistanceFromHome")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by DistanceFromHome")

# Show the plot
plt.show()
```

```python
df.groupby('YearsSinceLastPromotion')['Attrition'].mean().plot(kind='bar')

# Add labels and title
plt.xlabel("YearsSinceLastPromotion")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by YearsSinceLastPromotion")

# Show the plot
plt.show()
```

```python
df.groupby('WorkLifeBalance')['Attrition'].mean().plot(kind='bar')

# Add labels and title
plt.xlabel("WorkLifeBalance")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by WorkLifeBalance")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of 'Attrition'
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=df)
plt.title('Distribution of Attrition')
plt.show()

# Box plot to compare 'Age' for employees with and without attrition
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age Distribution for Attrition')
plt.show()
```

```python
# pie chart plot Attrition with Department
import matplotlib.pyplot as plt
df.groupby('Department')['Attrition'].mean().plot(kind='pie', autopct='%.2f%%')

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Department")

# Show the plot
plt.show()
```

```python
# pie chart plot Attrition with Age
import matplotlib.pyplot as plt
df.groupby('Age')['Attrition'].mean().plot(kind='pie', autopct='%.2f%%')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by Age")

# Show the plot
plt.show()
```

```python
# plot hist graph relation between attrition and age

df['Age'].hist(by=df['Attrition'], figsize=(10, 6))
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age by Attrition')
plt.show()
```

```python
# pie chart plot Attrition with Buisness Travel
import matplotlib.pyplot as plt
df.groupby('BusinessTravel')['Attrition'].mean().plot(kind='pie', autopct='%.2f%%')

# Add labels and title
plt.xlabel("BusinessTravel")
plt.ylabel("Attrition Rate")
plt.title("Attrition Rate by BusinessTravel")

# Show the plot
plt.show()
```
